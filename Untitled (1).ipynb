```python
import pandas as pd 
class DataPrepKit:
    def __init__(self,path):
       self.path=path
    def read_path(self):
        path_extension=self.path.split('.')[-1]
        if path_extension=='csv' or path_extension=='txt' or path_extension=='dat' or path_extension=='tsv':
            df=pd.read_csv(self.path)
        elif path_extension == 'xlsx' or path_extension == 'xls' or path_extension =='xlsm' or path_extension =='xlsb' or path_extension =='xltx' or path_extension =='xltm':
            df = pd.read_excel(self.path)
        elif path_extension =='json' or path_extension =='js' or path_extension =='jsonp':
            df=pd.read_json(self.path)
        return df
    def Data_Summary(self,df):
        description=df.describe() 
        print(description)
        data_shape = df.shape 
        print("Number of rows : ", data_shape[0])
        print("Number of Colmns : ", data_shape[1])
        data_types = df.dtypes
        print(data_types)
    def handle_Missing_Values(self,df):
        return df.fillna(0)
    def  Categorical_Data_Encoding(self,df):
        return pd.get_dummies(df)
       
##Replace this path with your path 
path1="C:\\Users\\MRC\\Downloads\\archive\\ds_salaries.csv"
obj=DataPrepKit(path1)
df=obj.read_path()
print(df)
print('Data Summary ')
obj.Data_Summary(df)
handle=obj.handle_Missing_Values(df)
print('Handle Missing Values :')
print(handle)
print('Categorical Encoding: ')
encoding=obj.Categorical_Data_Encoding(df)
print(encoding)
                                    
```
